```python
#importing needed libraries from python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#how to load dataset from local data
df=pd.read_excel("/content/Customer Call List.xlsx")
df
```

```python
df.info()
```

```python
#drop the duplicate row in the dataset
df.drop_duplicates(inplace=True)
df
```

```python
# Now remove unwanted symoblys from last name column
df['Last_Name']=df['Last_Name'].str.strip("/..._")
df
```

```python
#how to change integer datatype to string datatype
df['Phone_Number']=df['Phone_Number'].astype("str")
```

```python
#Now replace the phone number by using regex library
df['Phone_Number'] = df['Phone_Number'].str.replace(r'\D',"",regex=True)
df
```

```python
#Standard format.  - ----> 123-545-5421

s = "7066950392"
s[:3]+ "-" + s[3:6] + "-" + s[6:]


def st(s):
  return s[:3]+ "-" + s[3:6] + "-" + s[6:]
```

```python
#Now aggranging phone number column in proper format
df['Phone_Number'] = df['Phone_Number'].apply(lambda x:x[:3]+ "-" + x[3:6] + "-" + x[6:])
df
```

```python
#removing unwanted symbols from phone number column and also from dataset
df['Phone_Number'] = df['Phone_Number'].str.replace('--','')
df
```

```python
#In this dataset there are so many n/a values that's why we are replacing those values to nothing space
df.replace('N/a',"",inplace=True)
df
```

```python
#Now drop (not useful column) from dataset which is not useful for our dataset
df.drop(columns=['Not_Useful_Column'],inplace=True)
df
```

```python
#Replacing values yes to Y and no to N in paying customer column
df['Paying Customer']= df['Paying Customer'].str.replace("Yes","Y").str.replace("No","N").str.replace("N/a","")
```
